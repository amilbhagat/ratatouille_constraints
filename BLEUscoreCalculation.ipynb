```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
import pandas as pd

# Load the CSV file
csv_file_path = 'Combined_Recipes.csv'
csv_data = pd.read_csv(csv_file_path)

# Displaying the first few rows of the CSV file to understand its structure
csv_data.head()
```

```python
# Preprocessing and tokenizing the 'Recipe Instructions' column
# Lowercasing, removing punctuation, and tokenizing
csv_data['tokenized_instructions'] = csv_data['Recipe Instructions'].apply(lambda x: word_tokenize(x.lower()))

# Displaying the tokenized instructions for the first few recipes
csv_data['tokenized_instructions'].head()
```

```python
from nltk.translate.bleu_score import sentence_bleu

def calculate_bleu_scores(json_data, csv_data):
    bleu_scores = []
    for json_recipe, csv_recipe in zip(json_data, csv_data.iterrows()):
        reference = json_recipe['tokenized_instructions']  # Tokenized text from JSON
        candidate = csv_recipe[1]['tokenized_instructions']  # Tokenized text from CSV
        score = sentence_bleu([reference], candidate)
        bleu_scores.append(score)
    return bleu_scores
```

```python
import json
file_path = 'recipes.json'
with open(file_path, 'r') as file:
    json_data = json.load(file)
```

```python
import json
from nltk.tokenize import word_tokenize
import string

# Function to preprocess and tokenize text
def preprocess_and_tokenize(text):
    # Lowercase and remove punctuation
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize
    return word_tokenize(text)

# Preprocess and tokenize 'instructions' in each recipe
for recipe in json_data:
    # Joining all instructions into a single string, then tokenize
    combined_instructions = ' '.join(recipe['instructions'])
    recipe['tokenized_instructions'] = preprocess_and_tokenize(combined_instructions)

# Example: Print tokenized instructions of the first recipe
print(json_data[0]['tokenized_instructions'])
```

```python
# Calculate BLEU scores
bleu_scores = calculate_bleu_scores(json_data, csv_data)

# Print BLEU scores or calculate the average
print(bleu_scores)
average_bleu_score = sum(bleu_scores) / len(bleu_scores)
print(f"Average BLEU Score: {average_bleu_score}")
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# Function to calculate BLEU scores with smoothing
def calculate_bleu_scores_with_smoothing(json_data, csv_data):
    bleu_scores = []
    smoother = SmoothingFunction().method1  # Choose a smoothing method
    for json_recipe, csv_recipe in zip(json_data, csv_data.iterrows()):
        reference = json_recipe['tokenized_instructions']  # Tokenized text from JSON
        candidate = csv_recipe[1]['tokenized_instructions']
        score = sentence_bleu([reference], candidate, smoothing_function=smoother)
        bleu_scores.append(score)
    return bleu_scores

# Calculate BLEU scores with smoothing
bleu_scores = calculate_bleu_scores_with_smoothing(json_data, csv_data)

# Calculate the average
average_bleu_score = sum(bleu_scores) / len(bleu_scores)
print(f"Average BLEU Score with Smoothing: {average_bleu_score}")
```
